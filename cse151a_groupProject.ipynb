## Training Image Data Loader

```python
import matplotlib.pyplot as plt
import cv2
import os

#path to ISIC 2024 dataset on my google drive
google_drive_path ="/content/drive/MyDrive/cse151a-group-project/isic-2024-challenge/train-image/image"

#local path


i = 0
for filename in os.listdir(google_drive_path):
  if filename.lower().endswith(('.jpg')):
    file_path = os.path.join(google_drive_path,filename)

    image = cv2.imread(file_path)
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.show()
    i+=1
    if i==5:
      break
```
